# What are Operators?

👉 Operators are special symbols used to perform operations on variables and values

## Types of Operators in Python

### 1. Arithmetic Operators

👉 Used for mathematical calculations.

```python
a = 10
b = 3

print("Arithmetic Operators \n \n")

print("Addition:", a + b)          # 10 + 3 = 13
print("Subtraction:", a - b)       # 10 - 3 = 7
print("Multiplication:", a * b)    # 10 * 3 = 30
print("Division:", a / b)          # 10 / 3 = 3.333
print("Floor Division:", a // b)   # Removes decimal part -> 3
print("Modulus:", a % b)           # Remainder -> 1
print("Exponent:", a ** b)         # 10^3 = 1000
```

### 2. Comparison Operators

👉 Used to compare values.
👉 Output is always True or False.

```python
a=10
b=3

print("Comparison Operators\n \n")

print(a == b)   # False -> equal to
print(a != b)   # True  -> not equal to
print(a > b)    # True  -> greater than
print(a < b)    # False -> less than
print(a >= b)   # True  -> greater than or equal
print(a <= b)   # False -> less than or equal
```

### 3. Logical Operators

👉 Used to combine conditions.

```python
x = True
y = False

print("Logical Operators \n \n ")

print(x and y)   # True only if both are True
print(x or y)    # True if at least one is True
print(not x)     # Opposite value
```

### 4. Assignment Operators

👉 Used to assign values.

```python
num = 5

print("Assignment Operators \n \n")

num += 2   # num = num + 2
print(num)

num -= 1   # num = num - 1
print(num)

num *= 3   # num = num * 3
print(num)

num /= 2   # num = num / 2
print(num)
```

### 5.Membership Operators

👉 Used to check whether a value exists in a sequence.

```python
fruits = ["apple", "banana", "mango"]

print("Membership Operators \n")

print("apple" in fruits)        # True
print("grapes" not in fruits)   # True

```

### 6. Identity Operators

👉 Used to compare 

```python
x = 11
y = x

print(x)
print(y)

print("\n \n")
print(x is y)       # True
print(x is not y)   # False
```

### 7.Bitwise Operators

👉 Used to perform operations on binary numbers.

```python
p = 10   # Binary: 1010
q = 4    # Binary: 0100

print("Bitwise Operators\n")

print(p & q)   # AND
print(p | q)   # OR
print(p ^ q)   # XOR
print(~p)      # NOT
print(p << 1)  # Left shift
print(p >> 1)  # Right shift

```

## 1. if Statement

👉 Executes code only if condition is True

```python
age = 18

if age >= 18:
    print("Eligible for voting")
```

## 2. if-else Statement

👉 Executes one block if condition is true, otherwise another block.

```python
num = 5

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
num = 6

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
num = "99"

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

## 3. if-elif-else Statement

👉 Used to check multiple conditions.

```python
marks = 55

if marks >= 90:
    print("Grade A")
    
elif marks >= 70:
    print("Grade B")
    
elif marks>=50:
    print("Grade C")
    
else:
    print("Grade D")
```

## 4. Nested if Statement

👉 if inside another if

```python
username = "admin"
password = "1234"

if username == "admin":
    
    if password == "1234":
        print("Login Successful")
```

```python
username = "admin"
password = "12345"

if username == "admin":
    
    if password == "1234":
        print("Login Successful")
    else:
        print("condition 2 false")

else:
        print("condition 1 false")        
```

## 5. Short-hand if (Ternary Operator)

👉 One-line condition

```python
age = 20

result = "Adult" if age >= 18 else "Minor"

print(result)
```

# Most Asked Python Programs Using Conditional Statements

## 1. Even or Odd Number

```python
num = int(input("Enter number: "))

if num == 0:
    print("Zero")

elif num % 2 == 0:
    print("Even Number")

elif num % 2 != 0: 
    print("odd number") 
    
else: 
    print("enter valid input")
```

## 2. Positive, Negative, or Zero

```python
num = int(input("Enter number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
elif num == 0:
    print("Zero")
```

## 3. Largest of Two Numbers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("A is largest")
else:
    print("B is largest")
```

## 4. Largest of Three Numbers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a > b and a > c:
    print("A is largest")

elif b > c:
    print("B is largest")

else:
    print("C is largest")
```

## 5. Voting Eligibility

## 6. Simple Calculator

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

choice = input("Enter operator (+,-,*,/): ")

if choice == "+":
    print(a + b)

elif choice == "-":
    print(a - b)

elif choice == "*":
    print(a * b)

elif choice == "/":
    print(a / b)

else:
    print("Invalid Operator")
```
