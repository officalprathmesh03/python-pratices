```python
# Even or Odd Program
# Without Parameter
def even_odd():
    num = int(input("Enter a number: "))

    if num % 2 == 0:
        print("Even Number")
    else:
        print("Odd Number")

even_odd()

#With Parameter
def even_odd(num):
    if num % 2 == 0:
        print("Even Number")
    else:
        print("Odd Number")

even_odd(10)

# 2. Positive or Negative Program
# Without Parameter

def check_number():
    num = int(input("Enter a number: "))

    if num > 0:
        print("Positive Number")
    elif num < 0:
        print("Negative Number")
    else:
        print("Zero")

check_number()

#With Parameter

def check_number(num):
    if num > 0:
        print("Positive Number")
    elif num < 0:
        print("Negative Number")
    else:
        print("Zero")

check_number(-5)

# 3. Addition Program
# Without Parameter
def addition():
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))

    print("Sum =", a + b)

addition()


#With Parameter

def addition(a, b):
    print("Sum =", a + b)

addition(10, 20)

# 4. Subtraction Program
# Without Parameter

def subtraction():
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))

    print("Difference =", a - b)

subtraction()

#With Parameter

def subtraction(a, b):
    print("Difference =", a - b)

subtraction(20, 10)

# 5. Multiplication Program
# Without Parameter

def multiplication():
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))

    print("Product =", a * b)

multiplication()

# With Parameter

def multiplication(a, b):
    print("Product =", a * b)

multiplication(5, 4)

# 6. Division Program
# Without Parameter

def division():
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))

    print("Result =", a / b)

division()

# With Parameter

def division(a, b):
    print("Result =", a / b)

division(20, 5)
```
