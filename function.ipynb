A function in Python is created using the def keyword, followed by the function name and parentheses ().

Components of a Function:

• def → Keyword used to define a function.
• Function Name → Unique name of the function.
• Parameters → Inputs passed to the function (optional).
• Colon (:) → Marks the end of the function header.
• Docstring → Optional description of the function.
• Function Body → Statements that perform the task.
• Indentation → All statements inside the function must have the same indentation 
• return → Optional statement used to return a value from the function.

```python
def add_num(n1=10,n2=20):
    result = n1+n2
    return result

add_num()
     
```

# Function Scope in Python

it refers to the region of a program where a variable can be accessed.

    Types of Scope
1. Local Scope
2. Global Scope

## Local Scope
A variable created inside a function is called a Local Variable.

It can be accessed only inside that function.



```python
def student_details():

    student_name = "Rahul"
    print("inside fun--",student_name)
    print(student_name)
student_details()

#print(student_name)
```

```python
def student_details():

    student_name = "Rahul"

    print(student_name)

student_details()
```

## Global Scope
A variable created outside all functions is called a Global Variable.

It can be accessed anywhere in the program.(inside & outside the function)

```python
course_name = "Python"

def display_course():

    print("inside fun--",course_name)
    print("outside func--",course_name)
display_course()

print("inside fun--",course_name)
print("outside func--",course_name)
```

##  lambda function 

Syntax :

lambda arguments : expression



•	lambda is the keyword.
•	arguments are the input to the parameters (like in a normal function).(multiple arguments are allowed)
•	expression is evaluated and returned. (Only one expression allowed)

```python
add = lambda num1 , num2 : num1 + num2 
print(add(10,20))
```

## *args and **kwarges

### *args 
Used when the number of arguments is unknown.

*args collects all positional arguments into a tuple.



```python
def number (*args):
    print(args)
    
    print(sum(args))
number(10,20,30)
```

### **kwargs
Used when the number of keyword arguments is unknown.

**kwargs collects data into a dictionary.

```python
def student (**kwargs):
    print(kwargs)
student(name="Prathmesh",age=23)
```

# enumerate() Function in Python

enumerate() is used when you want both:

Index (position)
Value (element)

```python
Syntax:


enumerate(iterable)

or

enumerate(iterable, start)
```

```python
courses = ["Python", "SQL", "Power BI"]

for index, i in enumerate(courses):  # enumerate is return  index & value  

    print(index, i)
```
