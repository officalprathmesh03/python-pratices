# list 
A list is a mutable, ordered collection that allows duplicate elements. It is the most commonly used data
structure in Python.

## Characteristics
* Ordered — maintains insertion order
* Mutable — can change after creation
* Allows duplicates
* Supports any data type (heterogeneous)
* Indexed from 0 (also supports negative indexing)

```python
## Syntax
lst = [] 
print(lst)             # empty list
a = [1, 2, 3]
print(a)
b = ['a', 1, True, 3.14] 
print("mixed types",b)                        # mixed types
c = list(range(1, 6))  
print(c)                        # [1, 2, 3, 4, 5]
d = [x**2 for x in range(5)]
print("list comprehension",d)                         # list comprehension
```

## methods

### add element in list 

```python
# append() → Adds single element at end
# extend() → Adds multiple elements at end
# insert() → Adds single  element at specific position

a = [1,2,3,4]
print(a)

a.append(10)   # Add item in last of list 
print("Add item in last of list ",a)

a.extend([11,12,13])  # add multiple element in list
print("add multiple element in list",a)

a.insert(2,22) # add specific position 
print("add specific position ",a)
```

### remove element in list 

```python
a = [1,2,3,4,5,6,7,8]
print(a)

# pop()    → Removes last  single value
# remove() → Removes specific value/data
# del      → Removes element using index/position
# clear()  → Removes all elements

a.pop()   
print("Removes last  single value",a)

a.remove(4)
print("Removes specific value",a)

a.clear()
print("Removes all elements",a)
```

### counting and searching 

```python
# count()  -->  Count occurrences
# index()  -->  Find position/index of elements

a = [1,2,3,3,4,5,5,5,6,7,8]
print(a)

print ("count how much time repet ",a.count(3))

print ("find the index position  ",a.index(5))
```

```python
# reverse() --> reverse all the numbers
# sort()    --> sort ascending or descending (only list method)
# sorted()  --> sort ascending or descending (built in method)


a = [1,2,3,3,4,5,5,5,6,7,8]
print(a)

a.reverse()
print("reverse all the numbers",a)

a.sort()
print("sort ascending or descending",a)  # only for list 

sorted(a)
print("sort ascending or descending",a)

```

## Sum , All , Any 

```python
# sum() → Adds all elements
# all() → True if all values are True
# any() → True if at least one value is True

L1 = [1, 2, 3, 4, 5]

# sum()
print("Sum:", sum(L1))  #15


# all()
L2 = [True, True, False]
print("\nAll Elements True?:", all(L2))  #false

L22 = [True, True, True]
print("\nAll Elements True?:", all(L22)) #true



# any()
L3 = [False, False, True]
print("\nAt Least One True?:", any(L3))   #true

# split is used to seprated the words 
text = "apple,banana,mango"
print(text)
result = text.split(",")
print(result)

# join is used to jion the 2 statement 
words = ['Hello', 'World', 'Python']
print(words)
result = " ".join(words)
print(result)

# copy is used to copy the data 
a = [1,2,3,4,5,6]
b = a.copy()
print(a)
print("cpoy is used copy the varible ",b)
```
